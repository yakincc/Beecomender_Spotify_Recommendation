### Genre Analysis

```python
# Filter columns starting with 'genre'
genre_columns = [col for col in df.columns if col.startswith('genre')]

#Get the count of non zero values of each columnn
non_zero_counts = df[genre_columns].astype(bool).sum()

# Sort genre columns by count of non-zero values
sorted_columns = non_zero_counts.sort_values(ascending=False)
sorted_columns
```
